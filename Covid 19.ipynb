Comorbidities Analysis

```python
df.columns
```

```python
#  This shows which diseases are most common (likely Hypertension, Diabetes, and Obesity).

# List of comorbidity columns
comorbidities = ['DIABETES','COPD','ASTHMA','INMSUPR','HIPERTENSION',
                 'OTHER_DISEASE','CARDIOVASCULAR','OBESITY','RENAL_CHRONIC','TOBACCO']

# Count how many patients have each condition (1 = Yes)
condition_counts = {}
for col in comorbidities:
    condition_counts[col] = (df[col] == 1).sum()

# Convert to DataFrame for plotting
cond_df = pd.DataFrame.from_dict(condition_counts, orient='index', columns=['Count']).sort_values(by='Count', ascending=False)
print(cond_df)

# Plot bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=cond_df.index, y=cond_df['Count'], palette="viridis")
plt.title("Most Common Comorbidities in Patients")
plt.xticks(rotation=45)
plt.ylabel("Number of Patients with Condition")
plt.show()
```

```python
#  Link Between Comorbidities & Death
#This will reveal which diseases had the highest number of deaths.

# Death outcome column (if not created yet)
df['OUTCOME'] = df['DATE_DIED'].apply(lambda x: "Died" if x != "9999-99-99" else "Survived")

# Death counts per condition
death_counts = {}
for col in comorbidities:
    death_counts[col] = df[(df[col]==1) & (df['OUTCOME']=="Died")].shape[0]

death_df = pd.DataFrame.from_dict(death_counts, orient='index', columns=['Deaths']).sort_values(by='Deaths', ascending=False)
print(death_df)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=death_df.index, y=death_df['Deaths'], palette="rocket")
plt.title("Deaths Among Patients with Comorbidities")
plt.xticks(rotation=45)
plt.ylabel("Number of Deaths")
plt.show()
```

```python
#Death Rate (%) for Each Condition
#This shows which diseases are most strongly linked to death relative to their frequency.

# Death rate per condition
death_rate = {}
for col in comorbidities:
    total_with_condition = (df[col]==1).sum()
    deaths_with_condition = df[(df[col]==1) & (df['OUTCOME']=="Died")].shape[0]
    if total_with_condition > 0:
        death_rate[col] = (deaths_with_condition / total_with_condition) * 100

death_rate_df = pd.DataFrame.from_dict(death_rate, orient='index', columns=['Death Rate %']).sort_values(by='Death Rate %', ascending=False)
print(death_rate_df)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=death_rate_df.index, y=death_rate_df['Death Rate %'], palette="magma")
plt.title("Death Rate (%) by Comorbidity")
plt.xticks(rotation=45)
plt.ylabel("Death Rate (%)")
plt.show()
```

```python

# 1) Outcome from DATE_DIED (common coding: "9999-99-99" = survived)
df['OUTCOME'] = df['DATE_DIED'].apply(lambda x: "Died" if str(x) != "9999-99-99" else "Survived")

# 2) Make AGE_GROUP for easier analysis
bins  = [0, 18, 30, 50, 70, 120]
labels = ["0-18","19-30","31-50","51-70","71+"]
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=bins, labels=labels, right=True, include_lowest=True)

# 3) Helper to turn Yes/No coding into 1/0 (1=yes, 2=no; ignore 97/98/99 unknowns)
def yes_no(series):
    return series.map({1:1, 2:0})

# Quick peek
print(df[['SEX','AGE','OUTCOME','ICU','INTUBED','PREGNANT','TOBACCO','CLASIFFICATION_FINAL']].head())
```

1)  Hospitalization & ICU


A) How many needed ICU care? How many were intubated?

```python
icu_yes = (df['ICU'] == 1).sum()
intub_yes = (df['INTUBED'] == 1).sum()

print("Patients who needed ICU:", icu_yes)
print("Patients who were intubated:", intub_yes)

# Simple bar chart
counts = pd.Series({"ICU": icu_yes, "Intubated": intub_yes})
plt.figure(figsize=(5,4))
sns.barplot(x=counts.index, y=counts.values)
plt.title("ICU & Intubation Counts")
plt.ylabel("Number of Patients")
plt.show()
```

B) Which factors increased ICU admissions? (age & major conditions)

```python
# ICU rate by AGE_GROUP
icu_rate_by_age = (
    df[df['ICU'].isin([1,2])]
    .groupby('AGE_GROUP')['ICU']
    .apply(lambda s: (s==1).mean()*100)
    .sort_index()
)
print("ICU rate (%) by age group:")
print(icu_rate_by_age.round(2))

plt.figure(figsize=(7,4))
sns.barplot(x=icu_rate_by_age.index, y=icu_rate_by_age.values)
plt.title("ICU Rate (%) by Age Group")
plt.ylabel("ICU Rate (%)")
plt.xlabel("Age Group")
plt.show()

# ICU rate by key comorbidities
comorbidities = ['DIABETES','HIPERTENSION','OBESITY','CARDIOVASCULAR','RENAL_CHRONIC','INMSUPR']
icu_rates = {}
for col in comorbidities:
    sub = df[df[col].isin([1,2]) & df['ICU'].isin([1,2])]
    if len(sub)>0:
        icu_rate_with = (sub[sub[col]==1]['ICU']==1).mean()*100
        icu_rate_without = (sub[sub[col]==2]['ICU']==1).mean()*100
        icu_rates[col] = {"With": icu_rate_with, "Without": icu_rate_without}

icu_df = pd.DataFrame(icu_rates).T.sort_values("With", ascending=False)
print("\nICU rate (%) with vs without condition:")
print(icu_df.round(2))

# Plot ICU rate "With" the condition (easier to scan)
plt.figure(figsize=(8,4))
sns.barplot(x=icu_df.index, y=icu_df['With'])
plt.title("ICU Rate (%) Among Patients WITH Each Condition")
plt.xticks(rotation=45)
plt.ylabel("ICU Rate (%)")
plt.show()
```

2)  Pregnancy & COVID-19

Compare outcomes (death/survival) and ICU for pregnant vs non-pregnant women.

```python
#  2)  Pregnancy & COVID-19

# --- Death rate by pregnancy status ---
death_rate_preg = (
    df.groupby("PREGNANT", observed=True)["OUTCOME"]
      .apply(lambda x: (x == "Died").mean() * 100)
)

# --- ICU admission rate ---
icu_rate_preg = (
    df.groupby("PREGNANT", observed=True)["ICU"]
      .apply(lambda x: (x == 1).mean() * 100)
)

# --- Intubation rate ---
intub_rate_preg = (
    df.groupby("PREGNANT", observed=True)["INTUBED"]
      .apply(lambda x: (x == 1).mean() * 100)
)

#  Combine all into a summary DataFrame
rate_df = pd.concat(
    [death_rate_preg, icu_rate_preg, intub_rate_preg], 
    axis=1
)
rate_df.columns = ["Death Rate %", "ICU Rate %", "Intub Rate %"]
rate_df = rate_df.round(2)

# --- Print results ---
print("\nDeath Rate (%)  Pregnant vs Not Pregnant:\n", death_rate_preg.round(2))
print("\nICU Admission Rate (%)  Pregnant vs Not Pregnant:\n", icu_rate_preg.round(2))
print("\nIntubation Rate (%)  Pregnant vs Not Pregnant:\n", intub_rate_preg.round(2))
print("\n Summary Rates (%):\n", rate_df)

# --- Plot bar chart for better visualization ---
rate_df.plot(kind="bar", figsize=(7,5), rot=0, title="Pregnancy vs COVID-19 Outcomes (%)")
plt.ylabel("Percentage (%)")
plt.show()
```

3)  Lifestyle: Tobacco Use

Are smokers at higher risk of ICU or death?

```python
# Keep rows with known tobacco + known outcomes/ICU
tob = df[df['TOBACCO'].isin([1,2]) & df['ICU'].isin([1,2])]
tob['SmokeStatus'] = tob['TOBACCO'].map({1:'Smoker', 2:'Non-Smoker'})

# Death rate by smoking status
death_rate_smoke = tob.groupby('SmokeStatus').apply(lambda d: (d['OUTCOME']=="Died").mean()*100)
icu_rate_smoke = tob.groupby('SmokeStatus')['ICU'].apply(lambda s: (s==1).mean()*100)

print("Death Rate (%)  Smoker vs Non-Smoker:\n", death_rate_smoke.round(2))
print("\nICU Rate (%)  Smoker vs Non-Smoker:\n", icu_rate_smoke.round(2))

# Plot both
smoke_rates = pd.DataFrame({
    "Death Rate %": death_rate_smoke,
    "ICU Rate %": icu_rate_smoke
}).round(2)

smoke_rates.plot(kind='bar', figsize=(6,4))
plt.title("Smokers vs Non-Smokers  Death & ICU Rates")
plt.ylabel("Rate (%)")
plt.xticks(rotation=0)
plt.show()
```

4)  Classification of Cases (CLASIFFICATION_FINAL)

```python
# Make a simple CONFIRMED flag (assumption: 1/2/3 = confirmed)
df['CONFIRMED'] = df['CLASIFFICATION_FINAL'].apply(lambda x: 1 if x in [1,2,3] else 0)
print("Confirmed cases:", (df['CONFIRMED']==1).sum())
print("Not-confirmed/others:", (df['CONFIRMED']==0).sum())

# Outcomes by confirmation
out_by_conf = df.groupby(['CONFIRMED','OUTCOME']).size().unstack().fillna(0)
out_by_conf.index = out_by_conf.index.map({0:'Not Confirmed', 1:'Confirmed'})
print("\nOutcomes (counts):\n", out_by_conf)

# Death rate comparison
death_rate_conf = (
    df.groupby('CONFIRMED')
      .apply(lambda d: (d['OUTCOME']=="Died").mean()*100)
      .rename(index={0:'Not Confirmed', 1:'Confirmed'})
)
print("\nDeath Rate (%)  Confirmed vs Not Confirmed:\n", death_rate_conf.round(2))

# Plot
death_rate_conf.plot(kind='bar', figsize=(5,4))
plt.title("Death Rate (%)  Confirmed vs Not Confirmed")
plt.ylabel("Death Rate (%)")
plt.xticks(rotation=0)
plt.show()
```
